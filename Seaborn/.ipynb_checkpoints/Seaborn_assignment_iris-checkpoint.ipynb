```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
irisDataSet = sns.load_dataset('iris')
irisDataSet.head()
```

```python
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.title("Relational Plot")
plt.show()
```

```python
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', kind = 'line', data = irisDataSet)
plt.show()
```

```python
sns.set(style = "whitegrid")
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.show()
```

```python
sns.lmplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.title("LM Plot")
plt.show()
```

```python
sns.regplot( x = 'sepal_length',  y = 'sepal_width', data = irisDataSet)
plt.title("Regression Plot")
plt.show()
```

```python
correlation = irisDataSet.corr()
```

```python
sns.heatmap(correlation, annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
sns.kdeplot(x = 'sepal_length', data = irisDataSet, shade=True, color = "skyblue")
plt.title("KDE Plot")
plt.show()
```

```python
sns.kdeplot(x = 'sepal_width', data = irisDataSet, shade=True, color = "red")
plt.title("KDE Plot")
plt.show()
```

```python
sns.kdeplot(x = 'petal_length', data = irisDataSet, shade=True, color = "green")
plt.title("KDE Plot")
plt.show()
```

```python
sns.kdeplot(x = 'petal_width', data = irisDataSet, shade=True, color = "blue")
plt.title("KDE Plot")
plt.show()
```

```python
sns.displot(irisDataSet['sepal_length'])
plt.title("Distplot")
plt.show()
```

```python
sns.distplot(irisDataSet['sepal_width'], kde=False)
plt.title("Distplot")
plt.show()
```

```python
sns.countplot('species', data=irisDataSet)
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot('sepal_length', data=irisDataSet)
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot('sepal_width', data=irisDataSet)
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot('petal_length', data=irisDataSet)
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot('petal_width', data=irisDataSet)
plt.title("Count Plot")
plt.show()
```

```python
sns.countplot('petal_width', hue = 'species', data=irisDataSet)
plt.title("Count Plot")
plt.show()
```

```python
sns.boxplot('petal_width', 'species', data = irisDataSet , palette='PuRd')
plt.title("Box Plot")
plt.show()
```

```python
sns.boxplot('petal_width', 'species', data = irisDataSet , palette='rainbow')
plt.title("Box Plot")
plt.show()
```

```python
sns.violinplot('petal_width', 'species', data = irisDataSet , palette='rainbow')
plt.title("Violin Plot")
plt.show()
```

```python
sns.barplot('petal_width', 'species', data = irisDataSet , palette='rainbow')
plt.title("Bar Plot")
plt.show()
```
