```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python

sns.relplot(x='total_bill',y='tip',hue='time',data=tips)
```

## Lmplot()

```python
sns.lmplot(x='total_bill',y='tip',hue='time',data=tips)
```

## Regression plot

```python
plt.title('Regression line')
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
tips.head(10)
```

```python
corr = tips.corr() # pearson coeffient correlation #chi square correlation
```

```python
sns.heatmap(tips.corr(),annot=True,cmap=plt.cm.CMRmap_r)
```

## kdeplot (Kernal density plot)

```python
x = tips['total_bill']
sns.kdeplot(x,color='g',shade=True)
```

```python
x = tips['size']
sns.kdeplot(x,color='g',shade=True)
```

## distplot()

```python
x = tips['total_bill']
sns.distplot(x,kde=True)
```

```python
tips.head()
```

## Categoriacal data

```python
sns.countplot('time',data=tips)
```

## boxplot

```python
sns.boxplot('smoker','total_bill',data=tips,palette='PuRd')
```

```python
sns.boxplot('day','total_bill',data=tips,palette='PuRd')
```

```python
sns.boxplot('day','total_bill',hue='smoker',data=tips,palette='PuRd')
```

## Violineplot

```python

import seaborn
    
    
seaborn.set(style = 'whitegrid') 
tip = seaborn.load_dataset('tips')
  
seaborn.violinplot(x ='day', y ='tip', data = tip)
```
