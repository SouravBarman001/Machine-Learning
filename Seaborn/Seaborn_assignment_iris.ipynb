```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
irisDataSet = sns.load_dataset('iris')
irisDataSet.head()
```

```python
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.title("Relational Plot")
plt.show()
```

```python
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', kind = 'line', data = irisDataSet)
plt.show()
```

```python
sns.set(style = "whitegrid")
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.relplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.show()
```

```python
sns.lmplot( x = 'sepal_length',  y = 'sepal_width', hue='species', data = irisDataSet)
plt.title("LM Plot")
plt.show()
```

```python
sns.regplot( x = 'sepal_length',  y = 'sepal_width', data = irisDataSet)
plt.title("Regression Plot")
plt.show()
```

```python
correlation = irisDataSet.corr()
```

```python
sns.heatmap(correlation, annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
x = irisDataSet['sepal_length']
sns.kdeplot(x,shade=True, color = "skyblue")
plt.title("KDE Plot")
plt.show()
```

```python
x = irisDataSet['sepal_width']
sns.kdeplot(x,shade=True, color = "red")
plt.title("KDE Plot")
plt.show()
```

```python
x = irisDataSet['petal_length']
sns.kdeplot(x,shade=True, color = "green")
plt.title("KDE Plot")
plt.show()
```

```python
x = irisDataSet['petal_width']
sns.kdeplot(x,shade=True, color = "blue")
plt.title("KDE Plot")
plt.show()
```

```python
sns.distplot(irisDataSet['sepal_width'], kde=False)
plt.title("Distplot")
plt.show()
```
