```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
corr=titanic.corr()
```

```python
sns.heatmap(corr, annot=True, cmap=plt.cm.CMRmap_r)
```

```python
titanic.head()
```

```python
sns.countplot('age', data=titanic)
```

```python
sns.countplot('pclass', hue='class', data=titanic)
```

```python
sns.boxplot('pclass', 'age', data = titanic , palette='rainbow')
plt.title("Box Plot")
plt.show()
```

```python
sns.violinplot('age', 'class', data = titanic , palette='rainbow')
plt.title("Violin Plot")
plt.show()
```

```python
sns.distplot(titanic['age'])  # kde=False)
plt.title("Distplot")
plt.show()
```
