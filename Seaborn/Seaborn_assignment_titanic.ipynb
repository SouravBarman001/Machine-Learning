```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.set(style="darkgrid")
sns.relplot(x='fare',y='age', hue='sex', data=titanic)
plt.show()
```

```python
sns.set(style="white")
sns.relplot(x='fare',y='class', kind='line',  data=titanic)
plt.show()
```

# New Section

```python
sns.lmplot(x='fare',y='age',hue='sex', data=titanic)
```

```python
sns.regplot(x='fare', y='age', data=titanic)
plt.title("Rehreesion Line")
```

```python
corr=titanic.corr()
```

```python
sns.heatmap(corr, annot=True, cmap=plt.cm.CMRmap_r)
```

```python
sns.kdeplot(x='age', data=titanic, shade=True, color='yellow')
```
